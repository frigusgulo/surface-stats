```python
import georasters as gr
import numpy as np
import rasterio
```

```python
mckenzie_path = 'mckenzieDEM.tiff'
NDV, xsize, ysize, GeoT, Projection, DataType = gr.get_geo_info(mckenzie_path)
data = gr.from_file(mckenzie_path)
print(data.shape)
data.plot()
```

```python
# Make raster square, segment [400:1100,0:700]
mckenzie = gr.GeoRaster(data.raster[:,400:(400+data.shape[0])],data.geot,
                       nodata_value=data.nodata_value,
                       projection=data.projection,
                       datatype=data.datatype)
print(mckenzie.shape)
```

```python
mckenzie.plot()
mckenzie.to_tiff('mckenzie_mod')
```

```python
# Exploring rasterio module
mckenzie_io = rasterio.open('mckenzie_mod.tif')
print(mckenzio_io.crs)
def pixel2world(dem,x,y):
    # returns long/lat from pixel coordinates
    worldx, worldy = dem.transform*(x,y)
    return worldx,worldy
```

```python
pixel2world(mckenzie_io,0,0)
```
